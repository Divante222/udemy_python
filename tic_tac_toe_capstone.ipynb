```python
def board_update(game_board, user_choice, player_sign):
    if user_choice == '1' and game_board[5][0] == " ":
        game_board[5][0] = player_sign
    elif  user_choice == '2' and game_board[5][2] == " ":
        game_board[5][2] = player_sign
    elif  user_choice == '3' and game_board[5][4] == " ":
        game_board[5][4] = player_sign
        
    elif  user_choice == '4' and game_board[3][0] == " ":
        game_board[3][0] = player_sign
    elif  user_choice == '5' and game_board[3][2] == " ":
        game_board[3][2] = player_sign
    elif  user_choice == '6' and game_board[3][4] == " ":
        game_board[3][4] = player_sign  
        
    elif  user_choice == '7' and game_board[1][0] == " ":
        game_board[1][0] = player_sign 
    elif  user_choice == '8' and game_board[1][2] == " ":
        game_board[1][2] = player_sign 
    elif  user_choice == '9' and game_board[1][4] == " ":
        game_board[1][4] = player_sign 
    else:
        print('That spot has already been picked: ')
        return game_board, True
    return game_board, False
    
```
